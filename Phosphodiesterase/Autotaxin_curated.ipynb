```python
df_selected = df
f_smiles_not_splited = []
f_smiles_splited = []
no_need_to_split = []
selected_smiles = []
for i in df_selected['CANONICAL_SMILES']:
    if '.' in i:
        f_smiles_not_splited.append(i)
        f_smiles_splited.append(i.split('.'))
        splited = i.split('.')
        selected_smiles.append(max(splited,key=len))
    else:
        no_need_to_split.append(i)

df_selected.replace(f_smiles_not_splited,selected_smiles)
```
